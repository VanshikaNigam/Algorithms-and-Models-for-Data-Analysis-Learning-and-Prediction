```r
normalize <- function(x) {
    return ((x - min(x)) / (max(x) - min(x)))
  }
```

```r
german_data<-as.data.frame(german_data)
german_data_normed <- as.data.frame(lapply(german_data, normalize))
german_data_normed
```

```r
str(german_data_normed)
```

```r
german_data_normed<-as.data.frame(german_data_normed)
german_data_normed$Credit.Risks<-factor(german_data_normed$Credit.Risks, label=c("low","high"))
#german_data_subset=data.frame(german_data_subset)
head(german_data_normed)
```

```r
n.points=nrow(german_data_normed)
n.points
```

<b>80% Training Set ; 20% Test Set<b>

```r
set.seed(125)
sampling_rate_1 <-0.8
num.test.set.labels<- n.points* (1 - sampling_rate_1)

training<-sample(1:n.points, sampling_rate_1*n.points, replace=FALSE)
train_1<- subset(german_data_normed[training,],select=c(Age,Sex,Job,Housing,Saving.accounts,Checking.account,Credit.amount,Duration,Purpose))
#train
testing<-setdiff(1:n.points,training)
test_1<-subset(german_data_normed[testing,], select=c(Age,Sex,Job,Housing,Saving.accounts,Checking.account,Credit.amount,Duration,Purpose))
#test
cl_1<- german_data_normed$Credit.Risks[training]
#cl
true.labels<-german_data_normed$Credit.Risks[testing]
#true.labels
```

```r
dim(train_1)
dim(test_1)
length(cl_1)
```

```r
knn(train_1,test_1,cl_1,k=3) 
```

```r
#compute and print misclassification rate
for(k_1 in 1:20) {
  print(k_1)
  predicted.labels <- knn(train_1,test_1,cl_1,k_1)
  num.incorrect.labels<-sum(predicted.labels !=true.labels)
  misclassification.rate_1<-num.incorrect.labels/num.test.set.labels
  print(misclassification.rate_1)
    
}

```

The lowest Misclassification Rate when the training data is 80% and test data is 20% is  at <b>k=6</b> being equal to <b>0.255</b>

```r
knn(train_1,test_1,cl_1,k=6)
```

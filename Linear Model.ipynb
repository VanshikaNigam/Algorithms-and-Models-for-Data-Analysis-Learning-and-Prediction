<b>Part a) :</b> Evaluating the Default fit.

```r
fit_LR <- lm (NHL_DATA$A ~ NHL_DATA$G , data = NHL_DATA)
```

```r
summary(fit_LR)
attributes(fit_LR)
```

```r
plot(NHL_DATA$G ,NHL_DATA$A, pch = 16 ,cex=1.2, col="blue",xlab = "Goals",ylab="Assists",xlim = c(0,1000),ylim = c(0,2000))
abline(fit_LR,lwd=3,col="red")  
text(x  =500, y = 1500,labels="R_Squared = 0.1082\nP = 0.0008317")
```

<b>Part b) :</b> Changing the model so that the line is forced through Wayne Gretzky.

```r
goal=894
assist=1963
```

```r
fit_wayne<-lm(I(NHL_DATA$A-assist) ~ I(NHL_DATA$G-goal),data=NHL_DATA)
summary(fit_wayne)
coef(fit_wayne)
```

```r
intercept_wayne=-1003.42204909603
slope_wayne=0.520206294863515
```

```r
fit_wayne_final<-(lm(I(NHL_DATA$A-intercept_wayne)~I(NHL_DATA$G), data = NHL_DATA))
summary(fit_wayne_final)
```

```r
plot(NHL_DATA$G ,NHL_DATA$A, pch = 16 , cex=1.2,col="blue",xlab = "Goals",ylab="Assists",xlim = c(0,1000),ylim = c(0,2000))
abline(coef(fit_wayne_final) ,col="red",lwd=2)
text(x  =550, y = 1400,labels="R_Squared = 0.1082\nP = 0.0008317")
text(x=740,y=1980,labels="Wayne Gretzky")
```

<b>Part c) :</b> Changing the model so that the line is forced through Patrick Kane.

```r
kane_goal=285
kane_assist=462
```

```r
fit_kane<-lm(I(as.numeric(NHL_DATA$A)-kane_assist) ~ I(as.numeric(NHL_DATA$G)-kane_goal),data=NHL_DATA)
summary(fit_kane)
coef(fit_kane)
```

```r
intercept_kane=175.635760236941
slope_kane=0.537957578829376
```

```r
fit_kane_final<-(lm(I(as.numeric(NHL_DATA$A)-intercept_kane)~I(as.numeric(NHL_DATA$G)), data = NHL_DATA))
summary(fit_kane_final)
```

```r
plot(NHL_DATA$G ,NHL_DATA$A, pch = 16 ,cex=1.2, col="blue",xlab = "Goals",ylab="Assists",xlim = c(0,1000),ylim = c(0,2000))
abline(coef(fit_kane_final) ,col="red",lwd=2)
text(x  =500, y = 1500,labels="R_Squared = 0.116\nP = 0.0004909")
text(x=170,y=500,labels="Patrick Kane")
```

<b>Part d):</b> Fitting the above models through origin starts here!!

Making the Default Model go through (0,0)

```r
fit_LR_origin <- lm (I(as.numeric(NHL_DATA$A)) ~ I(as.numeric(NHL_DATA$G)) +0, data = NHL_DATA)
```

```r
summary(fit_LR_origin)
coef(fit_LR_origin)
```

```r
plot(NHL_DATA$G ,NHL_DATA$A, pch = 16 ,cex=1.2, col="blue",xlab = "Goals",ylab="Assists",xlim = c(0,1000),ylim = c(0,2000))
abline(fit_LR_origin,col="red",lwd=2)
text(x  =500, y = 1500,labels="R_Squared =0.8985\nP = 1.74679176345956e-51")
```

Making the Wayne Gretzky Model go through (0,0)

```r
fit_Wayne_origin<-lm(I(as.numeric(NHL_DATA$A)-assist) ~ I(as.numeric(NHL_DATA$G)-goal)+0,data=NHL_DATA)
```

```r
summary(fit_Wayne_origin)
coef(fit_Wayne_origin)
```

```r
slope_wayne_origin=assist/goal
```
